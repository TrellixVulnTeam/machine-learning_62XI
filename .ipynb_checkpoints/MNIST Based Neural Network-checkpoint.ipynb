```python
import numpy
import scipy.special

class neuralNetwork:
    def __init__(self, inputnodes, hiddennodes, outputnodes, learningrate):
        self.inodes = inputnodes
        self.hnodes = hiddennodes
        self.onodes = outputnodes
        
        # Init weights' maxtrices
        self.wih = numpy.random.normal(0.0, pow(self.hnodes, -0.5), (self.hnodes, self.inodes))
        self.who = numpy.random.normal(0.0, pow(self.onodes, -0.5), (self.onodes, self.hnodes))
        
        self.activation_function = lambda x: scipy.special.expit(x)
        
        self.lr = learningrate
        pass
    
    def train(self, inputs_list, targets_list):
        inputs = numpy.array(inputs_list, ndmin=2).T
        targets = numpy.array(targets_list, ndmin=2).T
        
        # See query part
        hidden_inputs = numpy.dot(self.wih, inputs)
        hidden_outputs = self.activation_function(hidden_inputs)
        final_inputs = numpy.dot(self.who, hidden_outputs)
        final_outputs = self.activation_function(final_inputs)
        
        # Calculate error
        output_errors = targets - final_outputs
        
        # Errors_hidden = Weights_T_hidden_output * Errors_ouput
        hidden_errors = numpy.dot(self.who.T, output_errors)
        
        # Update weights
        self.who += self.lr * numpy.dot((output_errors * final_outputs * (1.0 - final_outputs)), numpy.transpose(hidden_outputs))
        self.wih += self.lr * numpy.dot((hidden_errors * hidden_outputs * (1.0 - hidden_outputs)), numpy.transpose(inputs))
        pass
    
    def query(self, inputs_list):
        
        # Transform input:
        #
        #            [2]
        # [2,3,4] => [3]
        #            [4]
        inputs = numpy.array(inputs_list, ndmin=2).T
        
        # Hidden layer
        # X_hidden = W_input_hidden * I
        hidden_inputs = numpy.dot(self.wih, inputs)
        # O_hidden = sigmoid(X_hidden)
        hidden_outputs = self.activation_function(hidden_inputs)
        
        # Output layer
        final_inputs = numpy.dot(self.who, hidden_outputs)
        final_outputs = self.activation_function(final_inputs)
        return final_outputs
```

```python
from sklearn.datasets import fetch_mldata

mnist = fetch_mldata('MNIST original', transpose_data=True, data_home='datasets')
```

```python
X, y = mnist["data"], mnist["target"]
X.shape
```

```python
shuffle_index = numpy.random.permutation(60000)
X_train, X_test, y_train, y_test = X[shuffle_index], X[-shuffle_index], y[shuffle_index], y[-shuffle_index]
X_train
```

```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rc('axes', labelsize=14)
mpl.rc('xtick', labelsize=12)
mpl.rc('ytick', labelsize=12)

some_digit = X_train[3]
some_digit_image = some_digit.reshape(28, 28)
plt.imshow(some_digit_image, cmap = mpl.cm.binary,
           interpolation="nearest")
plt.axis("off")
plt.show()
y_train[3]
```

```python
input_nodes = len(X_train[0])
hidden_nodes = 200
output_nodes = 10

learning_rate = .1
n = neuralNetwork(input_nodes, hidden_nodes, output_nodes, learning_rate)

epochs = 5
for e in range(epochs):
    for index, item in enumerate(X_train):
        inputs = (numpy.asfarray(item) / 255 * 0.99) + 0.01
        targets = numpy.zeros(output_nodes) + 0.01
        targets[int(y_train[index])] = 0.99
        n.train(inputs, targets)
        pass
    print("Current training: ", e)
    pass
```

```python
fails = 0
fail_counter = numpy.zeros(output_nodes)

for index, item in enumerate(X_test):
    inputs = (numpy.asfarray(item) / 255 * 0.99) + 0.01
    outputs = n.query(inputs)
    predict = numpy.argmax(outputs)
    if (predict != y_test[index]):
        fails += 1
        fail_counter[int(y_test[index])] += 1
        print("Fail: Label ", y_test[index], " - predict ", predict)
        plt.imshow(item.reshape(28, 28), cmap = mpl.cm.binary,
                   interpolation="nearest")
        plt.axis("off")
        plt.show()
    pass
```

```python
print("Performance: ", (1 - (fails / len(X_test))) * 100, "%")
```

```python
for i in range(output_nodes):
    print("Guess value ", i, " wrong ", fail_counter[i], " times")
```
